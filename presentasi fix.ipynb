## Penjelasan Relasi One to One, Many to Many, dan One to Many pada Diagram Relasi Basis Data

Diagram relasi basis data yang Anda berikan menggambarkan hubungan antar tabel dengan menggunakan tiga jenis relasi utama: **one to one**, **many to many**, dan **one to many**. Berikut adalah penjelasan rinci mengenai relasi tersebut:

**1. Relasi One to One**

Relasi **one to one** (satu ke satu) terjadi ketika tepat satu baris data pada tabel pertama hanya berhubungan dengan tepat satu baris data pada tabel kedua. Pada diagram, relasi one to one terlihat pada tabel **gcr.type_forum** dan **gcr.forum_class**.

* Kolom `type_id` pada tabel **gcr.type_forum** terhubung dengan kolom `type_id` pada tabel **gcr.forum_class**.
* Setiap **type_forum** hanya dapat memiliki satu **forum_class**, dan setiap **forum_class** hanya dapat memiliki satu **type_forum**.

**Contoh:**

* Satu jenis forum (misalnya, forum diskusi) hanya dapat dikaitkan dengan satu kelas forum (misalnya, kelas pemrograman).
* Satu kelas forum tidak dapat dikaitkan dengan lebih dari satu jenis forum.

**2. Relasi Many to Many**

Relasi **many to many** (banyak ke banyak) terjadi ketika satu baris data pada tabel pertama dapat berhubungan dengan banyak baris data pada tabel kedua, dan sebaliknya. Pada diagram, relasi many to many terlihat pada tabel **gcr.users** dan **gcr.kelas**.

* Kolom `user_id` pada tabel **gcr.users** terhubung dengan kolom `user_id` pada tabel **gcr.joint_class**.
* Setiap **user** dapat mengikuti banyak **kelas**, dan setiap **kelas** dapat diikuti oleh banyak **user**.

**Contoh:**

* Satu siswa dapat mengikuti beberapa kelas (misalnya, matematika, fisika, dan kimia).
* Satu kelas dapat diikuti oleh beberapa siswa.

**3. Relasi One to Many**

Relasi **one to many** (satu ke banyak) terjadi ketika tepat satu baris data pada tabel pertama dapat berhubungan dengan banyak baris data pada tabel kedua, tetapi tidak sebaliknya. Pada diagram, terdapat beberapa relasi one to many:

* **gcr.forum_class** dan **gcr.assigment:** Satu forum kelas dapat memiliki banyak tugas, tetapi satu tugas hanya dapat dimiliki oleh satu forum kelas.
* **gcr.forum_class** dan **gcr.feedback:** Satu forum kelas dapat memiliki banyak umpan balik, tetapi satu umpan balik hanya dapat dimiliki oleh satu forum kelas.
* **gcr.users** dan **gcr.feedback:** Satu pengguna dapat memberikan banyak umpan balik, tetapi satu umpan balik hanya dapat diberikan oleh satu pengguna.
* **gcr.users** dan **gcr.teach_class:** Satu pengguna dapat mengajar banyak kelas, tetapi satu kelas hanya dapat diajar oleh satu pengguna.
* **gcr.assigment** dan **gcr.lampiran:** Satu tugas dapat memiliki banyak lampiran, tetapi satu lampiran hanya dapat dimiliki oleh satu tugas.

**Contoh:**

* Satu forum kelas dapat memiliki beberapa tugas, seperti tugas mingguan, tugas bulanan, dan tugas akhir.
* Satu tugas hanya dapat dimiliki oleh satu forum kelas.
* Satu forum kelas dapat memiliki banyak umpan balik dari siswa, pengajar, dan orang tua.
* Satu umpan balik hanya dapat dimiliki oleh satu forum kelas.
* Satu pengguna dapat memberikan banyak umpan balik untuk berbagai forum kelas dan tugas.
* Satu umpan balik hanya dapat diberikan oleh satu pengguna.
* Satu pengguna dapat mengajar beberapa kelas, seperti kelas matematika, fisika, dan kimia.
* Satu kelas hanya dapat diajar oleh satu pengguna.
* Satu tugas dapat memiliki beberapa lampiran, seperti dokumen, gambar, dan video.
* Satu lampiran hanya dapat dimiliki oleh satu tugas.

### Tentang Dataset (Manajemen Kelas)

Dataset ini mencakup berbagai tabel yang berkaitan dengan manajemen kelas di suatu institusi pendidikan. Berikut adalah deskripsi singkat tentang setiap tabel:

#### Tabel `users`
Tabel ini menyimpan informasi tentang pengguna sistem, seperti email, password, nama lengkap, nomor telepon, dan status aktif.

#### Tabel `assignments`
Tabel ini digunakan untuk melacak tugas yang diberikan kepada siswa. Setiap tugas memiliki informasi tentang waktu pembuatan, waktu pembaruan, forum yang terkait, penulis tugas, nilai, dan status tugas.

#### Tabel `feedback`
Tabel ini berisi feedback yang diberikan oleh pengguna terhadap tugas atau forum tertentu. Feedback dapat berupa komentar atau tanggapan terhadap tugas atau diskusi kelas.

#### Tabel `teach_class`
Tabel ini mencatat informasi tentang pengguna yang mengajar kelas tertentu. Setiap pengguna yang mengajar memiliki kode kelas yang terkait.

#### Tabel `type_forum`
Tabel ini menyimpan jenis-jenis forum yang dapat digunakan dalam kelas. Ini dapat mencakup jenis-jenis seperti diskusi, tugas, proyek, dan lain-lain.

#### Tabel `joint_class`
Tabel ini menunjukkan hubungan antara pengguna dan kelas yang diikuti. Ini memungkinkan pengguna untuk terdaftar dalam lebih dari satu kelas.

#### Tabel `forum_class`
Tabel ini digunakan untuk menyimpan informasi tentang forum yang diadakan dalam kelas. Setiap forum memiliki judul, deskripsi, batas waktu, dan jenis forum tertentu.

#### Tabel `lampiran`
Tabel ini berisi lampiran yang terkait dengan tugas atau forum tertentu. Lampiran ini bisa berupa file atau tautan yang diberikan oleh pengguna.

#### Tabel `kelas`
Tabel ini berisi informasi tentang kelas yang ada, termasuk kode kelas, nama kelas, dan status aktif kelas tersebut.

Dengan dataset ini, manajemen kelas dapat dilakukan secara efisien, memungkinkan untuk pelacakan tugas, forum diskusi, dan interaksi antara pengguna dengan mudah.

### Membuat table-table dari dataset yang sudah di normalisasi

### Tabel `users`

```python
%%sql
CREATE TABLE `users`(
    `user_id` BIGINT UNSIGNED NOT NULL AUTO_INCREMENT PRIMARY KEY,
    `email` VARCHAR(255) NOT NULL,
    `password` VARCHAR(255) NOT NULL,
    `is_active` TINYINT(1) NOT NULL,
    `full_name` VARCHAR(255) NOT NULL,
    `phonenumber` VARCHAR(255) NOT NULL
);
```

### Tabel `kelas`

```python
%%sql

CREATE TABLE `kelas`(
    `code_class` VARCHAR(255) NOT NULL,
    `name_class` VARCHAR(255) NOT NULL,
    `is_active` TINYINT(1) NOT NULL DEFAULT '1',
    PRIMARY KEY(`code_class`)
);
```

### Tabel `teach_class`

```python
%%sql
CREATE TABLE `teach_class`(
    `user_id` BIGINT UNSIGNED NOT NULL,
    `code_class` VARCHAR(255) NOT NULL,
    `id` BIGINT UNSIGNED NOT NULL AUTO_INCREMENT PRIMARY KEY,
    `creator` TINYINT(1) NOT NULL DEFAULT '0'
);
```

### Tabel `joint_class`

```python
%%sql
CREATE TABLE `joint_class`(
    `user_id` BIGINT UNSIGNED NOT NULL,
    `code_class` VARCHAR(255) NOT NULL,
    `id` BIGINT UNSIGNED NOT NULL AUTO_INCREMENT PRIMARY KEY
);
```

### Tabel `type_forum`

```python
%%sql
CREATE TABLE `type_forum`(
    `type_id` BIGINT UNSIGNED NOT NULL AUTO_INCREMENT PRIMARY KEY,
    `type_forum` VARCHAR(255) NOT NULL
);
```

### Tabel `forum_class`

```python
%%sql
CREATE TABLE `forum_class`(
    `forum_id` BIGINT UNSIGNED NOT NULL AUTO_INCREMENT PRIMARY KEY,
    `create_at` DATE NOT NULL,
    `update_at` DATE NOT NULL,
    `code_class` VARCHAR(255) NOT NULL,
    `deadline` DATETIME NULL,
    `title` VARCHAR(255) NOT NULL,
    `description` TEXT NOT NULL,
    `type_id` BIGINT UNSIGNED NOT NULL,
    `author` BIGINT UNSIGNED NOT NULL
);
```

### Tabel `assigment`

```python
%%sql
CREATE TABLE `assigment`(
    `assigment_id` BIGINT UNSIGNED NOT NULL AUTO_INCREMENT PRIMARY KEY,
    `create_at` DATETIME NOT NULL,
    `update_at` DATETIME NOT NULL,
    `forum_id` BIGINT UNSIGNED NOT NULL,
    `author` BIGINT UNSIGNED NOT NULL,
    `nilai` BIGINT NULL,
    `status` ENUM('ditugaskan', 'Telat','Selesai') NOT NULL DEFAULT 'ditugaskan'
);
```

### Tabel `feedback`

```python
%%sql
CREATE TABLE `feedback`(
    `feedback_id` BIGINT UNSIGNED NOT NULL AUTO_INCREMENT PRIMARY KEY,
    `author` BIGINT UNSIGNED NOT NULL,
    `create_at` DATETIME NOT NULL,
    `content` TEXT NOT NULL,
    `parent_feedback` BIGINT UNSIGNED NULL,
    `assigment_id` BIGINT UNSIGNED NOT NULL
);
```

### Tabel `lampiran`

```python
%%sql
CREATE TABLE `lampiran`(
    `lampiran_id` BIGINT UNSIGNED NOT NULL AUTO_INCREMENT PRIMARY KEY,
    `link_lampiran` VARCHAR(255) NOT NULL,
    `forum_id` BIGINT UNSIGNED NULL,
    `assigment_id` BIGINT UNSIGNED NULL
);
```

### CONTOH SELECT
* ketika ingin mengetahui berapa student pada setiap kelas


```python
%%sql
SELECT kelas.name_class, COUNT(users.user_id) as jumlah_siswa FROM users
JOIN joint_class ON users.user_id = joint_class.user_id
JOIN kelas ON joint_class.code_class = kelas.code_class
GROUP BY kelas.name_class;
```

### INNER JOIN


```python
%%sql 
select fc.code_class as kelas, fc.title as judul_forum, fc.description as deskripsi_forum ,tf.type_forum as jenis_forum,a.*, l.*
from forum_class fc
join type_forum tf on tf.type_id = fc.type_id
join assigment a on a.forum_id = fc.forum_id
join lampiran l on l.assigment_id = a.assigment_id or l.forum_id = fc.forum_id
limit 10;
```

#### RIGHT JOIN


```python
%%sql
SELECT k.name_class, tf.type_forum AS tipe_forum, fc.title as forum_title, fc.description AS deskripsi, fc.deadline, u.full_name AS nama_mahasiswa
FROM type_forum tf
RIGHT JOIN forum_class fc ON fc.type_id = tf.type_id
JOIN users u ON u.user_id = fc.author
JOIN kelas k ON k.code_class = fc.code_class
limit 10;
```

#### CROSS JOIN

```python
%%sql
SELECT tf.type_forum, k.name_class
FROM type_forum tf
CROSS JOIN kelas k
limit 10;
```
